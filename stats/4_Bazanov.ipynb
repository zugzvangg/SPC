# Вопросы

### №1 

Дайте определение систематической ошибки.

Систематическая ошибка - компонента погрешности, появляющаяся из-за искажения самой модели оценки, то есть bias. Она при каждом наблюдении завышает или занижает результат измерений, что зачастую связано с влиянием внешних факторов в ходе измерений.



### №2 

Как будут изменяться систематическая и статистическая ошибки с ростом статистики?

При росте статистики систематическая ошибка либо остается постоянной, либо меняется по закону, связанному с изменениями внешних факторов, из-за которых систематическая ошибка изначально и возникает. Статистическая же ошибка уменьшается пропорционально $1/\sqrt{N}$, $N$ - кол-во измерений

### №3 

Как можно улучшить систематические погрешности?

Так как первопричиной возникновения систематической погрешности являются внешние факторы, в первую очередь нужно попытаться исключить их из модели или хотя бы уменьшить их влияние. Если же первопричина изначально не понятна, можно попробовать менять условия эксперимента в целях найти модель, где систематическая ошибка не будет возникать, что позволит либо исключить её, либо хотя бы измерить её влияние в изначальной конфигурации эксперимента. Если же при изменении эксперимента систематическая ошибка не меняется, можно предположить, что все же это статистическая ошибка, и работать над ее уменьшением, увеличивая точность измерений. 

### №4 

Вы измерили что-то с помощью линейки, результат оказался между штрихам. Как вы оцените погрешность измерения? Почему?

Так как результат оказался именно между штрихами, а не на какой-то из отметок, нужно определять погрешность, используя тот факт, что величина измерения распределена по отрезку шириной в 1мм(цена деления линейки) равномерно.

Дисперсия равномерного распределения равна:

$$D = \frac{b - a}{12} = \frac{1}{12}$$

А значит: $$\sigma = \sqrt{D} = \frac{1}{2\sqrt{3}}$$

### №6

Вы произвели несколько измерений на установке и получили некоторый результат эксперимента: 
$x=42.0±1.3st±2.6sys$, где первое число это оценка, второе - статистическая ошибка, третье - систематическая. Имеет ли смысл дальше набирать статистику или стоит перейти к конструированию нового эксперимента?

В нашем эксперименте систематическая ошибка уже в 2 раза больше статистической. Если такая статистическая ошибка достаточна для нашего эксперимента, то можно больше не делать измерения и попробовать учесть систематическую ошибку, либо переходить к новому эксперименту и пытаться её снизить

# Задачи

### №1

Детектор зарегистрировал N событий, фон составил M событий. Какая статистическая погрешность?

Не фоновых событий всего: $L = N - M$. Фоновые и не фоновые события независимы, по сути мы имеем дело с Пуассоновским процессом.

Поэтому статистическая ошибка:

$$\sigma(L) = \sqrt{N - M}$$

### №2

Пусть вы провели измерение $x_1$ с ошибкой $\sigma$, потом вы провели второе такое измерение. Как ошибка среднего зависит от корреляции между измерениями?

То есть нужно оценить влияние среднего $(x_1 + x_2)/2$ от $\rho$ - корреляции

Дисперсия суммы зависимых величин записывается как:

$$D(x_1 + x_2) = D(x_1) + D(x_2) + 2\rho \sigma_1 \sigma_2$$

С учетом того, что:

$$\sigma_1 = \sigma_2 = \sigma $$

$$D(x_1 + x_2) = D(x_1) + D(x_2) + 2\rho \sigma^2 = \sigma^2 + \sigma^2 + 2\rho \sigma^2 = 2\sigma^2 + 2\rho \sigma^2 $$

Искомое:

$$D(\frac{x_1 + x_2}{2}) = \frac{D(x_1 + x_2)}{4} = \frac{2\sigma^2 + 2\rho \sigma^2}{4} = \frac{\sigma^2 + \rho \sigma^2}{2}$$

А значит искомая ошибка:

$$\sqrt{D(\frac{x_1 + x_2}{2})} = \sigma \sqrt{\frac{1 + \rho}{2}} $$

### №4

Исследуется асимметрия вперед-назад, эксперимент заканчивается после набора  $N$ событий, фиксированное число. Пусть $F$ и $B$ - числа в передней и задней полусферах соответственно $N=F+B$. По какому закону распределено 
$F$? Обычно асимметрия вперед-назад определяется как $r = \frac{F-B}{F+B}$ , оцените ее стандартное отклонение.

Так как N фиксировано и, если оно не слишком велико, можно сказать, что F распределено биномиально.

При обозначениях p = $\frac{F}{N}$ и $q = 1-p$:

Функция вероятности задается как:

$$\rho_F(k) = C_N^k p^k q^{N -k }$$

И функция распределения как:

$$ F_F(y) = \mathbb{P}(F <= y) = \sum_{k=0}^{[y]}C_N^k p^k q^{N -k } $$

Для биномиального распределения известно:

$$D(y) = Npq = N\frac{F}{N}\frac{N - F}{N} = \frac{F(N - F)}{N} $$

Тогда:

$$ \sigma(r) = \sqrt{D(r)} = \sqrt{D(\frac{F - B}{F + B})}  = 
\sqrt{D(\frac{F - B - F + F}{N})} = \sqrt{D(\frac{2F - N}{N})} = \sqrt{D(\frac{2F}{N} -1 )} = 
\sqrt{D(\frac{2F}{N})} = \sqrt{\frac{4}{N^2}D(F)} = \sqrt{\frac{4}{N^2}\frac{F(N - F)}{N}} = \frac{2}{N}\sqrt{\frac{F(N - F)}{N}}$$
